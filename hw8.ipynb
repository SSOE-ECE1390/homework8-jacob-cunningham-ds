```python
import cv2
import cv2.aruco as aruco
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Image paths
photo_path = os.path.relpath('hw8.jpg')
orig_path = os.path.relpath('Homework8.jpg')

# Load images as grayscale
photo = cv2.imread(photo_path, cv2.IMREAD_GRAYSCALE)
orig = cv2.imread(orig_path, cv2.IMREAD_GRAYSCALE)

# ArUco marker setup
a_dict = aruco.getPredefinedDictionary(aruco.DICT_4X4_1000)
a_parms = aruco.DetectorParameters()
a_detector = aruco.ArucoDetector(a_dict, a_parms)

# Detect markers
photo_corners, photo_ids, _ = a_detector.detectMarkers(photo)
orig_corners, orig_ids, _ = a_detector.detectMarkers(orig)

# Extract corner points
photo_pts = [corner[0][0] for corner in photo_corners[:3]]
orig_pts = [corner[0][0] for corner in orig_corners[:3]]

# Align photo to reference
transformation = cv2.getAffineTransform(np.float32(photo_pts), np.float32(orig_pts))
fixed_photo = cv2.warpAffine(photo, transformation, (orig.shape[1], orig.shape[0]))

# Plot results
fig, axs = plt.subplots(1, 3, layout='constrained')
axs[0].imshow(orig, cmap='gray')
axs[0].set_title('Given JPG')
axs[0].axis(False)

axs[1].imshow(photo, cmap='gray')
axs[1].set_title('Raw Photo')
axs[1].axis(False)

axs[2].imshow(fixed_photo, cmap='gray')
axs[2].set_title('Aligned Photo')
axs[2].axis(False)

plt.show()
```

```python
import easyocr

# Constants
dots_per_inch = 300
cm_to_pixels_conversion = dots_per_inch / 2.54  # Convert cm to pixels

# Function to crop a region from an image
def crop_region(image, region_width_cm, region_height_cm, region_offset_x_cm, region_offset_y_cm, cm_to_pixels):
    image_width_pixels = image.shape[1]
    image_height_pixels = image.shape[0]
    image_center_x = image_width_pixels // 2
    image_center_y = image_height_pixels // 2

    region_width_pixels = int(region_width_cm * cm_to_pixels)
    region_height_pixels = int(region_height_cm * cm_to_pixels)
    region_x_pixels = int(region_offset_x_cm * cm_to_pixels + image_center_x)
    region_y_pixels = int(region_offset_y_cm * cm_to_pixels + image_center_y)

    cropped_region = image[
        region_y_pixels:region_y_pixels + region_height_pixels,
        region_x_pixels:region_x_pixels + region_width_pixels
    ]
    return cropped_region

# Crop a region from the image
cropped_text_box = crop_region(fixed_photo, 12, 3, -11, -15, cm_to_pixels_conversion)

# Display the cropped region
plt.imshow(cropped_text_box, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import easyocr

# Initialize the OCR reader
ocr_reader = easyocr.Reader(['en'])

# Perform OCR on the cropped text box
ocr_results = ocr_reader.readtext(cropped_text_box)

# Print OCR results
print("OCR Results for the Text Box:")
for detected_text, text_content, confidence_score in ocr_results:
    print(f"Detected text: '{text_content}' with confidence: {confidence_score:.2f}")
```

```python
import pytesseract
import cv2
import matplotlib.pyplot as plt

# Set Tesseract command path
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# List of digit regions
digit_boxes = [digit_1_box, digit_2_box, digit_3_box, digit_4_box, digit_5_box]

# Function to preprocess each digit
def preprocess_digit(digit_box):
    # Normalize image to enhance contrast
    normalized_digit = cv2.normalize(digit_box, None, 0, 255, cv2.NORM_MINMAX)

    # Resize for better OCR accuracy
    resized_digit = cv2.resize(normalized_digit, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

    # Binarize using adaptive thresholding
    _, binary_digit = cv2.threshold(resized_digit, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    return binary_digit

# OCR on each digit
print("OCR Results:")
for digit_index, digit_box in enumerate(digit_boxes, start=1):
    # Preprocess the digit
    preprocessed_digit = preprocess_digit(digit_box)

    # Display the preprocessed digit for debugging
    plt.imshow(preprocessed_digit, cmap='gray')
    plt.title(f"Preprocessed Digit {digit_index}")
    plt.axis('off')
    plt.show()

    # Tesseract configuration for single character recognition
    config = "--psm 8 -c tessedit_char_whitelist=0123456789"

    # Perform OCR
    digit_text = pytesseract.image_to_string(preprocessed_digit, config=config).strip()
    print(f"Digit {digit_index} is: {digit_text}")

print("Note: The OCR results depend on preprocessing and may vary.")
```

```python
# Crop the region containing the face
face_box = crop_region(
    fixed_photo,
    region_width_cm=7.5,
    region_height_cm=10,
    region_offset_x_cm=3.25,
    region_offset_y_cm=-3,
    cm_to_pixels=cm_to_pixels_conversion
)

# Display the cropped face region
plt.imshow(face_box, cmap='gray')
plt.axis('off')
plt.title("Face Region")
plt.tight_layout()
plt.show()
```

```python
# Import MediaPipe for face recognition
import mediapipe as mp

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh_detector = mp_face_mesh.FaceMesh(
    static_image_mode=True, 
    max_num_faces=1, 
    refine_landmarks=True
)

# Convert the grayscale face box to RGB for MediaPipe processing
rgb_face_box = cv2.cvtColor(face_box, cv2.COLOR_GRAY2RGB)

# Detect facial landmarks
face_detection_results = face_mesh_detector.process(rgb_face_box)

# Draw landmarks on the face if detected
if face_detection_results.multi_face_landmarks:
    for detected_face in face_detection_results.multi_face_landmarks:
        for landmark_id, landmark in enumerate(detected_face.landmark):
            image_height, image_width = rgb_face_box.shape[:2]
            x_coord, y_coord = int(landmark.x * image_width), int(landmark.y * image_height)
            cv2.circle(rgb_face_box, (x_coord, y_coord), radius=2, color=(0, 255, 0), thickness=-1)

# Display the face region with landmarks
plt.imshow(rgb_face_box)
plt.axis('off')
plt.title("Facial Landmarks")
plt.tight_layout()
plt.show()
```
